17. Describe the time complexity of accessing elements in a dictionary.
  - | Operation                   | Average Case | Worst Case |
| --------------------------- | ------------ | ---------- |
| Access (`dict[key]`)        | O(1)         | O(n)       |
| Insertion (`dict[key]=val`) | O(1)         | O(n)       |
| Deletion (`del dict[key]`)  | O(1)         | O(n)       |
  1. Average Case O(1)

    - Hash function distributes keys evenly → one direct index → constant time.

  2. Worst Case O(n)

    - If many keys hash to the same index (collision), Python handles it with a technique like open addressing.
    - In theory, you may need to scan all items → linear time.
    - In practice, this is extremely rare due to Python’s good hash function.
