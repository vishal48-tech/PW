```python
# 9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50
# and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a
# histogram to visualize the distribution
import numpy as np
import matplotlib.pyplot as plt

rand_1000 = np.random.normal(50, 5, 1000)
print(np.mean(rand_1000))
print(np.std(rand_1000))
fig, ax = plt.subplots()
ax.hist(rand_1000, bins=10, edgecolor = "black")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
ax.set_title("Histogram")
plt.show()
```

```python
# 10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
# 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
# Write the Python code to compute the mean sales and its confidence interval.

# Answer
# According to the CLT, if we take many samples of size n from a population:
# - The sampling distribution of sample means will approximate a normal distribution.

import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Step 1: Compute sample statistics
n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation

# Step 2: Set confidence level
confidence = 0.95
alpha = 1 - confidence
z_critical = stats.norm.ppf(1 - alpha/2)  # 1.96 for 95%

# Step 3: Compute margin of error
margin_of_error = z_critical * (std_sales / np.sqrt(n))

# Step 4: Compute confidence interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print(f"Sample Mean: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
