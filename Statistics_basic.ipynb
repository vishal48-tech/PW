1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.
  - Descriptive statistics summarizes the complete data.
    - Eg. Rohit Sharma's performance in current tournament.
  - Inferential statistics summarizes the sample of complete data.
    - Eg. Calculating number of trees in forest using 10000 trees / sq. ft.

2. What is sampling in statistics? Explain the differences between random and stratified sampling.
  - In statistics, sampling means collecting some amount of data from complete data.
    - Random sampling means taking data from any random person.
    - Stratified sampling means creating equal groups and taking data from any random person of each group.

3. Define mean, median, and mode. Explain why these measures of central tendency are important.
  - The arithmetic mid value of data is called mean.
  - The physical mid value of data is called median.
  - The highest frequency value of data is called mode.
  - It is important because it gives the central representation of data.

4. Explain skewness and kurtosis. What does a positive skew imply about the data?
  - Skweness is the measure of asymmetry of probability distribution. 0 means symmetry, else asymmetry.
  - Kurtosis is the measure of tailedness of probability distribution.
  - Positive skew imply that more data is on left and right has tail.

```python
# 5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
import statistics as stat
import numpy as np

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

print(np.mean(numbers), np.median(numbers), stat.mode(numbers))
```

```python
# 6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

print(np.cov(list_x, list_y))
print(np.corrcoef(list_x, list_y))
```

```python
# 7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.show()

# Line before box is lower fence
# Lower box line is Q1
# Inside box line is Q2
# Upper box line is Q3
# Line after box is upper fence
# Data point above upper fence is outlier
```

```python
# 8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# ● Explain how you would use covariance and correlation to explore this relationship.
# ● Write Python code to compute the correlation between the two lists:

# Answer:
# Using covariance i can tell if advertising_spend increases, daily_sales increases and
# correlation tells how much advertising_spend affects daily_sales.

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

print(np.corrcoef(advertising_spend, daily_sales))
```
